#### Grab & Upload to Azure & Extract Data

```python
# Azure Functions
def azure_upload_blob(connect_str, container_name, blob_name, data):
    blob_service_client = BlobServiceClient.from_connection_string(connect_str)
    blob_client = blob_service_client.get_blob_client(container=container_name, blob=blob_name)
    blob_client.upload_blob(data, overwrite=True)
    print(f"Uploaded to Azure Blob: {blob_name}")

def azure_download_blob(connect_str, container_name, blob_name):
    blob_service_client = BlobServiceClient.from_connection_string(connect_str)
    blob_client = blob_service_client.get_blob_client(container=container_name, blob=blob_name)
    download_stream = blob_client.download_blob()
    return download_stream.readall()
```

```python
# Specify the path to your JSON configuration file
config_file_path = 'config.json'

# Load the JSON configuration file
with open(config_file_path, 'r') as config_file:
    config = json.load(config_file)

# Print the configuration
#Connection_STRING = config["connectionString"]

CONNECTION_STRING_AZURE_STORAGE = config["connectionString"]
CONTAINER_AZURE = 'budgetexpense'

# Initialize the BlobServiceClient
blob_service_client = BlobServiceClient.from_connection_string(CONNECTION_STRING_AZURE_STORAGE)

# Get the container client
container_client = blob_service_client.get_container_client(CONTAINER_AZURE)


budget_expense_df = pd.DataFrame()

# List all blobs in the specified container
blob_list = container_client.list_blobs()
for blob in blob_list:
    print(blob.name)
    blob_client = container_client.get_blob_client(blob=blob.name)
    blob_data = blob_client.download_blob()
    blob_content = blob_data.readall().decode('utf-8')
    df = pd.read_csv(StringIO(blob_content))
    # Display the head of the DataFrame
    print(df.shape)
    budget_expense_df = df.copy()# Copy the DataFrame to the main DataFrame
```

```python
##adding 'Budget_deficit' col
budget_expense_clean['Budget_deficit'] = budget_expense_clean['Budget'] - budget_expense_clean['Expenditures']
budget_expense_clean
```

```python
# Create Fact table

# Add Dept_rollup foreign key
fact_table = pd.merge(budget_expense_clean, df_Dept_rollup, left_on='Dept_rollup_name', right_on='Dept_rollup_name', how='left')

# Add Department foreign key
fact_table = pd.merge(fact_table, df_Department, left_on='DepartmentName', right_on='DepartmentName', how='left')

# Add Fund foreign key
fact_table = pd.merge(fact_table, df_Fund, left_on='FundName', right_on='FundName', how='left')

# Add Program foreign key
fact_table = pd.merge(fact_table, df_Program, left_on='ProgramName', right_on='ProgramName', how='left')

# Add Activity foreign key
fact_table = pd.merge(fact_table, df_Activity, left_on='ActivityName', right_on='ActivityName', how='left')

# Add Unit foreign key
fact_table = pd.merge(fact_table, df_Unit, left_on='UnitName', right_on='UnitName', how='left')

# Add Expense foreign key
fact_table = pd.merge(fact_table, df_Expense, left_on='ExpenseName', right_on='ExpenseName', how='left')


# Add 'fact_id' column before creating the fact table
fact_table['fact_id'] = range(1, len(fact_table) + 1)

# Create the fact table
fact_table = fact_table[['fact_id', 'Budget_fiscal_year', 'Thru_quarter','Budget_deficit', 'Dept_rollup_name', 'DepartmentName', 'FundName', 'ProgramName', 'ActivityName', 'UnitName', 'ExpenseName', 'Key', ]]

fact_table
```
